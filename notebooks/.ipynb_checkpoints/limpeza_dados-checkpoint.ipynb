# Limpeza dos Datasets sobre Vacinação

Nesse notebook realizaremos a limpeza dos datasets originais obtidos no site do DATASUS.

# Bibliotecas

```python
import pandas as pd
```

# Dataset 1: Tipo de Vacina por Região

```python
tipo_regiao_raw = pd.read_csv('../dados/raw/tipo_vacina_por_regiao_1994_2019.csv',
                                sep = ';',
                                encoding = 'ISO-8859-1',
                                skiprows = 3,
                                skipfooter = 20,
                                thousands = '.',
                                decimal = ',',
                                engine = 'python')
```

```python
tipo_regiao_raw
```

```python
tipo_regiao = tipo_regiao_raw.copy()
```

```python
# remove os 3 números do nome do imunizante
tipo_regiao['Imuno'] = tipo_regiao['Imuno'].str[4:]
```

```python
tipo_regiao.columns = ['Imuno', 'Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste', 'Total']
```

```python
tipo_regiao.info()
```

```python
# salvar em um novo csv
tipo_regiao.to_csv(r'../dados/clean/tipo_regiao.csv')
```

# Dataset 3: Tipo de Vacina por Ano

```python
tipo_ano_raw = pd.read_csv('../dados/raw/tipo_vacina_por_ano.csv',
                                sep = ';',
                                encoding = 'ISO-8859-1',
                                skiprows = 3,
                                skipfooter = 20,
                                thousands = '.',
                                decimal = ',',
                                engine = 'python')
```

```python
tipo_ano_raw.head()
```

```python
tipo_ano = tipo_ano_raw.copy()
```

```python
tipo_ano['Imuno'] = tipo_ano['Imuno'].str[4:]
```

```python
tipo_ano = tipo_ano.replace('-', 0)
```

```python
tipo_ano.info()
```

```python
colunas = tipo_ano.columns[1:] # tira a coluna 'Imuno' da conversão
tipo_ano[colunas] = tipo_ano[colunas].apply(pd.to_numeric, errors = 'coerce')
```

```python
tipo_ano.info()
```

```python
tipo_ano.to_csv(r'../dados/clean/tipo_ano.csv')
```

# Dataset 4: Quantidade de Pessoas Imunizadas por Região do País

```python
imunizados_regiao_raw = pd.read_csv('../dados/raw/imunizacao_regiao_1994_2019.csv',
                                sep = ';',
                                encoding = 'ISO-8859-1',
                                skiprows = 3,
                                skipfooter = 20,
                                thousands = '.',
                                decimal = ',',
                                engine = 'python')
```

```python
imunizados_regiao_raw
```

Como podemos ver, o dataset já está relativamente bem organizado, faremos apenas alguns ajustes.

```python
imunizados_regiao = imunizados_regiao_raw.copy()
```

```python
imunizados_regiao = imunizados_regiao.set_index('Ano').rename_axis(None)
```

```python
imunizados_regiao.columns = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste', 'Total']
```

```python
imunizados_regiao.info()
```

```python
imunizados_regiao.to_csv(r'../dados/clean/imunizados_regiao.csv')
```
